```python
#importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
#Opening my spotify_list
spotipy_df = pd.read_csv ('df_w_audio_ftotal.csv')
spotipy_df
```

```python
#Selecting the data to the cluster process
float_columns = spotipy_df.select_dtypes(include=['number'])
spotipy_df1 = spotipy_df[float_columns.columns]
spotipy_df1
```

```python
# Despite of the values seem to be scaled, some of them are negative, so let's scale them to be sure they are in the same range
from sklearn.preprocessing import StandardScaler
X_prep= StandardScaler().fit_transform(spotipy_df1)
transformer = StandardScaler().fit(spotipy_df1)
# Now, all features will have the same weight.
pd.DataFrame(X_prep,columns=spotipy_df1.columns).head()
```

# Choosing k

```python
# INERTIA GRAPH
from sklearn.cluster import KMeans
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
'''By taking a look in to the inertia graph, we see that the 'elbow shape' is around 9 and 10'''
```

```python
# SILHOUETTE GRAPH
from sklearn.metrics import silhouette_score
K = range(2, 20)# generally a good range of values to see the behaviour
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_prep)
    silhouette.append(silhouette_score(X_prep, kmeans.predict(X_prep)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

```python
'''By taking a look in to the silhouette graph, despite of the highest peak be around 2, it doesn't look to be good enough
 number of cluster so I will consider th highest second peak that is around 8.

Acordingly to both graphs obtained i can conlude that the best number of clusters is between 8 and 10 and for that reason I will
choose to use 9 clusters.
'''
```

# Building the Model

```python
# Using the number of clusters concluded before(9)
kmeans = KMeans(n_clusters = 9, random_state=42).fit(X_prep)
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_prep)
# Checking the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
# Plotting the clusters
plt.figure(figsize=(5, 5))
plt.scatter(X_prep[:, 0], X_prep[:, 9], c=clusters,alpha=0.5) 
plt.show()  
```

```python
# Saving the clusters and the trasnformer for future scalling
import pickle
#saving kmeans_clusters
pickle.dump(kmeans, open('kmeans.pkl', 'wb'))
#kmeans = pickle.load(open('kmean.pkl', 'rb')) 
#saving scaler
pickle.dump(transformer, open('scaler.pkl', 'wb'))
#scaler = pickle.load(open('X_prep.pkl','rb'))
```

```python
# Adding the column 'cluster' to my dataframe
spotipy_df ['cluster'] = clusters
```

```python
# Getting my final dataframe with the cluster column
spotify_df_w_clusters = spotipy_df[['track.name', 'name', 'cluster']]
spotify_df_w_clusters
```

```python
# Saving my last dataframe wih the clusters
spotify_df_w_clusters.to_csv('spotify_df_w_clusters.csv', index=False)
```

# Creating an updated function to song recomendation

```python
#importing the necessary libraries
from spotipy.oauth2 import SpotifyClientCredentials
import spotipy
from pandas import json_normalize
```

```python
#oppening my popvortex 100 hot songs file
popv_hot_100 = pd.read_csv('top_songs.csv')
popv_hot_100.head()
```

```python
# Stablishing  spotify connection
def open_spotify_connection():
    secrets_file = open("Secrets","r")
    string = secrets_file.read()
    secrets_dict={}
    for line in string.split('\n'):
        if len(line) > 0:
            #print(line.split(':'))
            secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
    sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
    return sp;
```

```python
# Building a song recommendation model function 
def song_recommender(title):
    # Convert the user input and song names to lowercase for case-insensitive comparison
    user_input_song_lower = user_input_song.lower()
    top_chart_songs_lower = popv_hot_100.apply(lambda x: x.str.lower())
    # Check if the user-inputted song is in the top chart songs list (case-insensitive)
    if user_input_song_lower in top_chart_songs_lower['song_title'].values:
        # Filter out the user-inputted song from the DataFrame
        top_chart_songs_without_input = popv_hot_100[top_chart_songs_lower['song_title'] != user_input_song_lower]
        # Recommend a random song from the filtered DataFrame
        recommended_song = top_chart_songs_without_input.sample().squeeze()
        print(f"That's a Hot song! You might also like: {recommended_song['song_title']} by {recommended_song['artist_name']}")
    else:
        #open spotify connection
        sp = open_spotify_connection()
        # Query Spotify
        results = sp.search(q=f"track:{user_input_song_lower}", type="track", limit=1)
        # Check if the query returns results (i.e., total > 0)
        if results['tracks']['total'] != 0:
            track_id = results['tracks']['items'][0]['id']
            #print(track_id)
            new_data = sp.audio_features(track_id)
            data_df = json_normalize(new_data[0])
            num_data_df = data_df.select_dtypes(include='number')
            # Use pickle to get the existing scaler and kmeans 
            kmeans = pickle.load(open('kmeans.pkl', 'rb'))
            scaler = pickle.load(open('scaler.pkl', 'rb'))
            scaler = scaler.transform(num_data_df)
            predicted_cluster = kmeans.predict(scaler)
            cluster = predicted_cluster[0]
            #print (cluster)
            rec_song = spotify_df_w_clusters[spotify_df_w_clusters['cluster'] == cluster].sample().squeeze()
            print(f"That's in my spotify_list!You might also like: {rec_song['track.name']} by {rec_song['name']}")
        else:
            print("The song is not in Spotify. Please try again")
```
